# 算出「強勢股」「弱勢股」
* 把台股加權指數當作基準指數(benchmark index)，以個股作為比較指數，比較同一個時期內的累積漲跌幅，就可以算出個股是否比大盤強或弱。

## 步驟一、先計算出台股加權指數20天的漲跌幅%

2019/3/4 台股加權指數收盤是10349.88點，3/29加權指數收盤是10541.43點
加權指數20天的漲跌幅%= ((10541.43-10349.88)/ 10349.88)x100= + 1.85%

```python
print(df_TWII['Adj Close'])

# print(df.loc[:,'Adj Close'])
```

```python
# 假設你已經有一個名為 twii_close 的列表，包含了每天的收盤價
twii_close = df_TWII['Close']

window_size = 20  # 滾動窗口大小，即 20 天
twii_start_prices = []  # 用於存放每個滾動窗口的起始價格
twii_end_prices = []    # 用於存放每個滾動窗口的結束價格
twii_20days_returns = [] # 用於存放每個滾動窗口的漲跌幅

for i in range(len(twii_close) - window_size + 1):
    # 提取每個滾動窗口的起始價格和結束價格
    twii_start_price = twii_close[i]
    twii_end_price = twii_close[i + window_size - 1]
    
    twii_start_prices.append(twii_start_price)
    twii_end_prices.append(twii_end_price)

# 顯示結果
for start_price, end_price in zip(twii_start_prices, twii_end_prices):
#     print(f"Start Price: {start_price}, End Price: {end_price}")
    
    twii_20days_return = ((end_price - start_price) / start_price) * 100
    twii_20days_returns.append(twii_20days_return)
    print("台股加權指數 20 天的漲跌幅%:", twii_20days_return)
```

## 步驟二、再算出台積電(2330)20天的漲跌幅%

2019/3/4 台積電收盤價是235.5元，2019/3/29台積電收盤價是245.5元
台積電20天的漲跌幅%=((245.5-235.5)/235.5)x100= + 4.25%

```python
# 假設你已經有一個名為 t2300_close 的列表，包含了每天的收盤價
t2300_close = df_2330['Close']

window_size = 20  # 滾動窗口大小，即 20 天
t2300_start_prices = []  # 用於存放每個滾動窗口的起始價格
t2300_end_prices = []    # 用於存放每個滾動窗口的結束價格
t2300_20days_returns = [] # 用於存放每個滾動窗口的漲跌幅

for i in range(len(twii_close) - window_size + 1):
    # 提取每個滾動窗口的起始價格和結束價格
    t2300_start_price = t2300_close[i]
    t2300_end_price = t2300_close[i + window_size - 1]
    
    t2300_start_prices.append(t2300_start_price)
    t2300_end_prices.append(t2300_end_price)

# 顯示結果
for start_price, end_price in zip(t2300_start_prices, t2300_end_prices):
#     print(f"Start Price: {start_price}, End Price: {end_price}")
    
    t2300_20days_return = ((end_price - start_price) / start_price) * 100
    t2300_20days_returns.append(t2300_20days_return)
    print("台積電(2330) 20 天的漲跌幅%:", t2300_20days_return)
```

## 步驟三、計算台積電強(弱)度值

(4.25 – 1.85)/1.85= + 1.297
由以上實例，我們可以算出台積電(2330)的強(弱)度值就是 + 1.297

```python
strength_value = (t2300_20days_return - twii_20days_return) / twii_20days_return
print("台積電(2330) 的強(弱)度值:", strength_value)
```

```python
# 步驟三、計算台積電(2330) 的強(弱)度值
strength_values = [(t2300_return - twii_return) / twii_return for t2300_return, twii_return in zip(t2300_20days_returns, twii_20days_returns)]
print("台積電(2330) 的強(弱)度值:", strength_values)
```

```python
import random
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta

y_twii = "^TWII"
y_2330 = "2330.TW"

startdate = datetime(2019, 1, 1)  # 訓練起始日期
enddate = datetime(2020, 1, 1)  # 訓練結束日期

df_TWII = yf.download(y_twii, start=startdate, end=enddate)
df_2330 = yf.download(y_2330, start=startdate, end=enddate)

# 步驟一、計算台股加權指數 20 天的漲跌幅%
twii_close = df_TWII['Close']
window_size = 20  # 滾動窗口大小，即 20 天
twii_start_prices = []  # 用於存放每個滾動窗口的起始價格
twii_end_prices = []    # 用於存放每個滾動窗口的結束價格

for i in range(len(twii_close) - window_size + 1):
    # 提取每個滾動窗口的起始價格和結束價格
    twii_start_price = twii_close[i]
    twii_end_price = twii_close[i + window_size - 1]
    
    twii_start_prices.append(twii_start_price)
    twii_end_prices.append(twii_end_price)

# 印出每個滾動窗口的結果
for start_price, end_price in zip(twii_start_prices, twii_end_prices):
    twii_20days_return = ((end_price - start_price) / start_price) * 100
#     print("台股加權指數 20 天的漲跌幅%:", twii_20days_return)

# 步驟二、計算台積電(2330) 20 天的漲跌幅%
t2300_close = df_2330['Close']
t2300_start_prices = []  # 用於存放每個滾動窗口的起始價格
t2300_end_prices = []    # 用於存放每個滾動窗口的結束價格

for i in range(len(t2300_close) - window_size + 1):
    # 提取每個滾動窗口的起始價格和結束價格
    t2300_start_price = t2300_close[i]
    t2300_end_price = t2300_close[i + window_size - 1]
    
    t2300_start_prices.append(t2300_start_price)
    t2300_end_prices.append(t2300_end_price)

# 印出每個滾動窗口的結果
for start_price, end_price in zip(t2300_start_prices, t2300_end_prices):
    t2300_20days_return = ((end_price - start_price) / start_price) * 100
#     print("台積電 20 天的漲跌幅%:", t2300_20days_return)

# 步驟三、計算台積電(2330) 的強(弱)度值
strength_values = [(t2300_20days_return - twii_return) / twii_return for t2300_20days_return, twii_return in zip(t2300_20days_returns, twii_20days_returns)]
print("台積電(2330) 的強(弱)度值:", strength_values)
```

## 弱勢股舉例


晶電(2448)20天的強(弱)度值
步驟一、計算出台股加權指數20天的漲跌幅%是+ 1.85%
前面已經計算過故不再贅述。
步驟二、再算出晶電(2448)20天的漲跌幅%
2019/3/4 晶電收盤價是27.05元，2019/3/29晶電收盤價是24.8元
晶電20天的漲跌幅%=((24.8-27.05)/27.05)x100= - 8.32%
步驟三、計算晶電強(弱)度值
( - 8.32 –1.85)/1.85= - 5.5
由以上實例，我們可以算出晶電(2448)的強(弱)度值是 – 5.497

所以很明顯的比較出台積電(2330)是強勢股而晶電(2448)是弱勢股
這裡面的20天是中期強度，若要計算短期強度則可以把20天改為5天，
若要計算長期強度則可以把20天改為120天計算。
